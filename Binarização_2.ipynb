```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# %matplotlib inline


```

```python
imagem = mpimg.imread('calhetas.jpg.jpg')
imagem = np.array(imagem)
plt.imshow(imagem)

```

```python
imagem_tons_de_cinza = 0.2989*imagem[:,:,0] + 0.5870*imagem[:,:,1] + 0.1140*imagem[:,:,2]
plt.imshow(imagem_tons_de_cinza, cmap='gray')
```

```python
imagem_binaria = imagem_tons_de_cinza > 127
plt.imshow(imagem_binaria, cmap='gray')
```

```python
plt.subplot(1, 3, 1)
plt.axis('off')
plt.imshow(imagem)
plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(imagem_tons_de_cinza, cmap='gray')
plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(imagem_binaria, cmap='gray')
plt.show()
```
